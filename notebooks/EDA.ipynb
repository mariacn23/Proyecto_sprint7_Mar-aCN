```python
import pandas as pd
import plotly.express as px

# Desde notebooks/ sube un nivel a la raíz
df = pd.read_csv("../vehicles_us.csv")

# Vista rápida
df.head()

df.shape, df.dtypes

# Para evitar problemas con nulos en las columnas que graficaremos
df_plot = df.dropna(subset=["odometer", "price"])

fig_1 = px.histogram(df_plot, x="odometer", nbins=50,
                   title="Distribución del odómetro")
fig_1

```

```python
fig_2 = px.scatter(df_plot, x="odometer", y="price", opacity=0.5,
                 title="Precio vs Odómetro", trendline=None)
fig_2
```
